Passando os valores para variáveis:

```python
# 19/01
qtd_total_vendas_1 = 3
ticket_medio_1 = 320.52

# 20/01
valor_total_vendas_2 = 834.47
ticket_medio_2 = 119.21

# 23/01
valor_total_vendas_3 = 15378.12
qtd_total_vendas_3 = 5
```

Calculando o Ticket Médio:

```python
A = valor_total_vendas_1 = qtd_total_vendas_1 * ticket_medio_1
B = qtd_total_vendas_2 = valor_total_vendas_2 / ticket_medio_2
C = ticket_medio_3 = valor_total_vendas_3 / qtd_total_vendas_3
```

Mostrando os resultados:

```python
print(f'19/01: {A:.2f}, 20/01: {B:.2f}, 23/01: {C:.2f}')
```

# Extraindo da string abaixo o valor da taxa selic e dos anos.

```python
noticia = 'Selic vai a 2,75% e supera expectativas é a primeira alta em 6 anos.'
```

```python
posicao_final = noticia.find('%')

posicao_inical = noticia.rfind(' ', 0, posicao_final) +1

selic = noticia[posicao_inical:posicao_final]

ano = noticia[noticia.find("6")]
```

Mostrando os resultados:

```python
print(selic)
print(ano)
```
